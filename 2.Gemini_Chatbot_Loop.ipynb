```python
from google import genai
gemini=genai.Client(api_key='API')
```

```python
system=input("What type of chatbot would you like to create?\n")
print('Your new assistant is ready!')

while True:    
    user=input()
    if user.strip().lower() =='quit()':
        print('Chatbot session Ended.')
        break
    response=gemini.models.generate_content(model='gemini-2.0-flash-lite',contents=user)
    print("\n" + response.text + "\n")
```
